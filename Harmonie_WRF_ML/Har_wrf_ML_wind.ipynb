# Wind speed

```python
#@title Wind intensity metar statistic
result["spd_o"].describe()
```

```python
#@title velocity intensity Harmonie
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import median_absolute_error
import seaborn as sns
import numpy as np

#Mean absolute error
Mean_AE = round(mean_absolute_error(result["spd_o"], result["mod0_h"]),2)
#MSE 
MSE = round(mean_squared_error(result["spd_o"], result["mod0_h"]),2)
#R2
R2 = round(r2_score(result["spd_o"], result["mod0_h"]),2)
#Median absolute error
Median_AE = round(median_absolute_error(result["spd_o"], result["mod0_h"]),2)
#random seed 1
np.random.seed(1)

sns.set_theme(style="darkgrid")
g = sns.jointplot(x="mod0_h", y="spd_o", data=result.sample(200),
                  kind="reg", truncate=False,
                  xlim=(0, 15), ylim=(0, 15),
                  color="m", height=7)


print(pd.DataFrame({"Mean_AE":Mean_AE,"MSE":MSE,"R2":R2,"Median_AE":Median_AE},index=["results point 0"]))
```

# Wind direction

```python
#@title wind direction metars
#wind direction to interval dir=-1 variable wind
interval = pd.IntervalIndex.from_tuples([(-1.5, -0.5),(-0.5,20), (20, 40), (40, 60),
                                           (60,80),(80,100),(100,120),(120,140),(140,160),
                                           (160,180),(180,200),(200,220),(220,240),
                                           (240,260),(260,280),(280,300),(300,320),
                                           (320,340),(340,360)])

labels = ['VRB', '[0, 20]', '(20, 40]', '(40, 60]','(60, 80]', '(80, 100]',
          '(100, 120]', '(120, 140]','(140, 160]', '(160, 180]', '(180, 200]',
          '(200, 220]','(220, 240]', '(240, 260]', '(260, 280]', '(280, 300]',
          '(300, 320]', '(320, 340]', '(340, 360]']

result["dir_l"] = pd.cut(result["dir_o"], bins=interval,retbins=False,
                        labels=labels).map({a:b for a,b in zip(interval,labels)}).astype('category')


#observed climatology
print("wind direction climatology\n")
print(result["dir_l"].value_counts(normalize= True).transform(lambda x: '{:,.0%}'.format(x)))

result["spd_o_kt"]=np.rint(result["spd_o"]*1.94384)
print("\nwind intensity percentage (kt) when wind variable\n")
print(result[result.dir_l=="VRB"]["spd_o_kt"].value_counts(normalize = True).transform(lambda x: '{:,.0%}'.format(x)))
print("\nwind direction when wind speed equal or less than 3 kt\n")
result[result.spd_o_kt<=3]["dir_l"].value_counts(normalize = True).transform(lambda x: '{:,.0%}'.format(x))
```

```python
#@title Harmonie. Quality report
df_novrb = result[result.dir_o!=-1]

from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns


#label meteorological model variables dir0 and dir1
df_novrb.loc[:,"dir0_l"] = pd.cut(df_novrb["dir0_h"], bins=interval,retbins=False,labels=labels).map({a:b for a,b in zip(interval,labels)}).astype('category')
df_novrb.loc[:,"dir1_l"] = pd.cut(df_novrb["dir1_h"], bins=interval,retbins=False,labels=labels).map({a:b for a,b in zip(interval,labels)}).astype('category')

plt.figure(figsize=(12, 9))
plt.title("Quality report meteorologic model point 0")
cr = pd.DataFrame(classification_report(df_novrb.dir_l, df_novrb.dir0_l,output_dict=True)).T
sns.heatmap(cr[:-3].drop(columns=["support"]),annot=True,cmap="YlGnBu",fmt='.0%');
```

```python
#@title Harmonie global quality report
plt.figure(figsize=(8, 6))
plt.title("Global quality report (meteorologic model point 0)")
sns.heatmap(cr[-3:].drop(columns=["support"]),annot=True,cmap="YlGnBu",fmt='.1%');
```

```python
#@title Harmonie. Confusion matrix
plt.figure(figsize=(12, 9))
plt.title("Confusion matrix (meteorologic model point 0)")

global_sc = pd.crosstab(df_novrb.dir_l, df_novrb.dir0_l,margins=True,)
sns.heatmap(global_sc,annot=True,cmap="YlGnBu",fmt='.0f');
```

```python
#@title Harmonie. Precision and entropy meteorologic model point 0
import math
from scipy.stats import entropy


plt.figure(figsize=(12, 9))
plt.title("Precision and entropy meteorologic model point 0")
column_sc = pd.crosstab(df_novrb.dir_l, df_novrb.dir0_l, margins=True,normalize="columns")
column_sc = column_sc.append(pd.DataFrame(entropy(column_sc,base=2)/(math.log2(column_sc.shape[0])),columns=["entropy/entropy.max"],
                    index=column_sc.columns).T).rename(columns={"All":"Climatology"})
sns.heatmap(column_sc,annot=True,cmap="YlGnBu",fmt='.0%');
```
